```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# Definimos una función para graficar imágenes
def graficar_imagenes(imagenes, etiquetas, num_imagenes=10):
    plt.figure(figsize=(10, 4))
    for i in range(num_imagenes):
        plt.subplot(2, 5, i + 1)  # Crear una cuadrícula de 2 filas y 5 columnas
        plt.imshow(imagenes[i], cmap='gray')  # Mostrar la imagen en escala de grises
        plt.title(f'Etiqueta: {etiquetas[i]}')  # Mostrar la etiqueta correspondiente
        plt.axis('off')  # Ocultar los ejes
    plt.show()
```

1. Preparamos el conjunto de datos

```python
# Cargamos el conjunto de datos MNIST
mnist = tf.keras.datasets.mnist

# Dividimos el conjunto de datos en entrenamiento y prueba
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalizamos los datos a un rango de 0 a 1
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Llamamos a la función para graficar las primeras 10 imágenes
graficar_imagenes(x_train, y_train)
```

2. Creación de la red neuronal

```python
# Definimos nuestro modelo como secuencial con capas densas
modelo = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Aplanamos las imágenes de 28x28 a un vector de 784
    tf.keras.layers.Dense(units=128, activation='relu'),  # Capa oculta con 128 neuronas y activación ReLU
    tf.keras.layers.Dense(units=10, activation='softmax')  # Capa de salida con 10 neuronas para clasificación
])
```
