```python
# Importamos las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def graficar(eje_x, eje_y, ):
    plt.plot(eje_x, eje_y, color='red', label='Datos reales')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Regresión Lineal')
    plt.legend()
    plt.show()
```

Utilizaremos la ecuación y = 2x + 5 como base
Utilizaremos una semilla 1 para obtener en cada ejecución los mismos números aleatorios  

Ecuación de la recta:
y = mx + b  
donde:
- y: Es el valor de la coordenada "𝑦" para un punto cualquiera de la recta.
- x: Es el valor de la coordenada "𝑥" para un punto cualquiera de la recta.
- m: La pendiente indica la inclinación de la recta.

```python
# Fijar la semilla
np.random.seed(1)

# Generamos nuestro set de datos y definimos nuestra ecuación para nuestros datos de entrenamiento
x = np.random.randint(0, 100, size=100).reshape(-1, 1)
y = 2 * x + 5

# Graficamos los valores reales
graficar(x, y)
```

```python
# Creamos el modelo de regresión lineal y lo entrenamos
modelo = LinearRegression()
modelo.fit(x, y)

# Realizamos predicciones
y_pred = modelo.predict(x)

for eje_y, eje_y_p in zip(y, y_pred):
    print(f'Valor real: {eje_y}, Valor predicción: {eje_y_p}')
```

```python
# Visualizamos los resultados
plt.plot(x, y, color='red', label='Datos reales')
plt.plot(x, y_pred, color='green', label='Predicción del modelo')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal: y = 2x + 5')
plt.legend()
plt.show()

# Imprimimos los coeficientes del modelo
print(f"Coeficiente (pendiente): {modelo.coef_[0][0]}")
print(f"Intersección (ordenada al origen): {modelo.intercept_[0]}")
```
